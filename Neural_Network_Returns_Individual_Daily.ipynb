```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (7, 7)
import pandas as pd
import yfinance as yf
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta

import random
import math
```

# Data Cleaning

```python
Start_Date_Str='2010-07-01'
```

```python
data=yf.download('TSLA ',start=Start_Date_Str,interval='1d')
TSLA_Rend=data['Close'].pct_change()
```

```python
data
```

```python
TSLA_Rend= pd.DataFrame(TSLA_Rend)
TSLA_Rend
```

```python
window_size = 16
transformed_TSLA_Rend = []

# Building the observations, 15 predictors returns and the change predicted
for i in range(1,len(TSLA_Rend) - window_size+1):
    row = TSLA_Rend['TSLA'][i:i + window_size].values
    transformed_TSLA_Rend.append(row)

# Create a DataFrame
columns = [f'Change_{i}' for i in range(1, window_size)] + ['ChangeToPredict']
transformed_TSLA_Rend = pd.DataFrame(transformed_TSLA_Rend, columns=columns)

# Results
transformed_TSLA_Rend
```

```python
ChangeToPredict=transformed_TSLA_Rend['ChangeToPredict']
ChangeToPredict
```

```python
del transformed_TSLA_Rend['ChangeToPredict']
transformed_TSLA_Rend
```

# Neural Network

```python
from keras import Sequential
from keras.layers import Dense, Embedding, LSTM, GRU
from keras.layers import Dropout, LeakyReLU
from keras.callbacks import EarlyStopping
from keras.layers import BatchNormalization, LayerNormalization
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.optimizers import Adam
```

```python
Top=int(ChangeToPredict.size*.8)+1
```

```python
Y_Train=ChangeToPredict[0:Top]
Y_Test=ChangeToPredict[Top:]
X_Train=transformed_TSLA_Rend[0:Top]
X_Test=transformed_TSLA_Rend[Top:]
```

```python
modeloUni = Sequential()
modeloUni.add(Dense(64, input_dim=15))  # Número de entradas o features
modeloUni.add(LeakyReLU(alpha=0.01))  # Usar Leaky ReLU
modeloUni.add(Dense(50))
modeloUni.add(LeakyReLU(alpha=0.01))
modeloUni.add(Dropout(0.25))
modeloUni.add(Dense(32))
modeloUni.add(LeakyReLU(alpha=0.01))
modeloUni.add(Dense(1, activation='linear'))
modeloUni.summary()
```

```python
random.seed(996)
Batch_size = 25
modeloUni.compile(loss='mean_squared_error', optimizer='adam', metrics=["mean_squared_error"])
modeloUni.fit(X_Train, Y_Train, epochs=500, batch_size=Batch_size, verbose=1);
```

```python
modeloUni.evaluate(X_Test, Y_Test)
```
